## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
import pandas as pd

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['cut','price']]

grafico =  sns.barplot(data=diamonds, x="cut", y="price", ci=None, palette="pastel")
```



**Insight do gráfico 1**: Podemos ver que o preço do Very Good e o Good São bastante semelhante

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head(13)
```

```
# gráfico do exercício 2

flights = data.query('month == "Dec"')
with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x="year", y="passengers", palette="pastel")
  grafico.set(title='Passageiros por ano em dezembro', xlabel='Ano', ylabel='Passageiros');
```

**Insight do gráfico 2**: Nesse grafico pode se perceber que o maior salto de numeros de passageiros no mês de dezembro foi de 1958 a 1959.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3
flights = data.query("1949 <= year <= 1959")

with sns.axes_style('whitegrid'):

  altura  = 25 / 2.54
  largura = 25 / 2.54
  plt.figure(figsize=(largura,altura))
  grafico = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")

  grafico.set(title='Passageiros por mês entre 1949 a 1959', xlabel='Mês', ylabel='Passageiros');
  grafico.get_legend().set_title("Ano");
```

**Insight do gráfico 3**: com o passar dos anos a tendência é aumentar o número de passageiros e sempre do mês de julho à agosto ocorre padrões com o alto números de passageiros.
